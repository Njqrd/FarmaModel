```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Simulate a DataFrame with monthly data for 5 years


df = pd.read_csv('sp500.csv')
df['Date'] = pd.to_datetime(df['Date'], utc=True)
# Step 2: Calculate monthly returns
df['MONTHLY_RETURN'] = df['Close'].pct_change()

# Step 3: Determine quarterly returns
df['QUARTER'] = df['Date'].dt.to_period('Q')
quarterly_returns = df.groupby('QUARTER')['MONTHLY_RETURN'].sum()

# Step 4: Identify recession periods (two consecutive quarters with negative growth)
recession_quarters = quarterly_returns[quarterly_returns < 0].index
recession_flags = recession_quarters.to_series().diff(1) == 1
recession_periods = recession_quarters[recession_flags]

# Convert recession periods to date range for plotting
recession_dates = []
for period in recession_periods:
    start_date = pd.Period(period, freq='Q').start_time
    end_date = pd.Period(period, freq='Q').end_time
    recession_dates.append((start_date, end_date))

# Step 5: Calculate cumulative returns
df['CUMULATIVE_RETURN'] = (1 + df['MONTHLY_RETURN']).cumprod() - 1

# Step 6: Plotting
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['CUMULATIVE_RETURN'], label='Cumulative Return')

# Highlight recession periods
for start_date, end_date in recession_dates:
    plt.axvspan(start_date, end_date, color='red', alpha=0.3)

plt.title('S&P 500 Cumulative Returns with Recession Periods Highlighted')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data_path = 'ReturnsS&P.csv'
try:
    data = pd.read_csv(data_path,parse_dates={'Date': ['Year', 'Month']})
    data['Date'] = pd.to_datetime(data['Date'])
    data.set_index('Date', inplace=True)
    data.index = data.index.to_period('M')  # Convert DateTimeIndex to PeriodIndex for monthly data
except Exception as e:
    print(f"Error loading data: {e}")

# Ensure 'Return' is numeric
data['Return (%)'] = pd.to_numeric(data['Return (%)'], errors='coerce')

# Identify recession periods function
def identify_recessions(df, column='Return (%)'):
    recession_periods = []
    in_recession = False
    start_date = None

    for date, row in df.iterrows():
        if row[column] < 0:
            if not in_recession:
                start_date = date
                in_recession = True
        else:
            if in_recession:
                end_date = date
                # Ensure at least two months of negative returns to count as a recession
                # For monthly data, directly compare the months difference
                if (end_date.year - start_date.year) * 12 + (end_date.month - start_date.month) > 1:
                    # Decrement the end date by one month for the recession period end
                    if end_date.month == 1:  # Handle January case
                        recession_end = end_date
                    else:
                        recession_end = end_date
                    recession_periods.append((start_date, recession_end))
                    in_recession = False

    # Check for ongoing recession at the end of data
    if in_recession:
        recession_periods.append((start_date, df.index[-1]))

    return recession_periods


recession_ranges = identify_recessions(data)

# Calculate cumulative returns
cumulative_returns = (1 + data['Return (%)'] / 100).cumprod() - 1



# Plotting
plt.figure(figsize=(14, 7))
plt.plot(cumulative_returns, label='Cumulative Monthly Returns', color='blue')

# Highlight recession periods
for start, end in recession_ranges:
    plt.axvspan(start, end, color='red', alpha=0.3)

plt.title('Cumulative Monthly Returns with Recession Periods Highlighted')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.grid(True)
plt.show()
```
